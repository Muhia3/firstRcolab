```r
# Set the seed for reproducibility
set.seed(123)

# Number of samples
n_samples <- 10000

# Rate parameter for the exponential distribution
rate_parameter <- 3

# Step 1: Generate uniform random variables
u <- runif(n_samples)

# Step 2: Apply the inverse transform
exponential_samples <- -log(1 - u) / rate_parameter

# Save the generated values for later use
write.table(exponential_samples, "exp_samples.txt", col.names = FALSE)

# Display the first few values
head(exponential_samples)
```

```r
# Read the generated exponential samples from the file
exponential_samples <- read.table("exp_samples.txt", header = FALSE)[, 1]

# Number of samples in each mean calculation
sample_size <- 10

# Calculate means of samples
means <- tapply(exponential_samples, rep(1:(n_samples/sample_size), each = sample_size), mean)

# Plot histogram on the probability density scale
hist(means, prob = TRUE, col = "skyblue", main = "Histogram of Sample Means",
     xlab = "Sample Means", ylab = "Probability Density")

# Add a curve for the theoretical density of the mean of Exp(3)
curve(dnorm(x, mean = 1/3, sd = sqrt(1/(3*sample_size))), add = TRUE, col = "red", lwd = 2)

# Add a legend
legend("topright", legend = c("Empirical", "Theoretical"), fill = c("skyblue", "red"))
```
